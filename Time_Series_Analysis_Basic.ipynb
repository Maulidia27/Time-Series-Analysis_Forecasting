# **TASK 1**

```python
# simple line chart
df1.plot(
    x = 'index', 
    y = 'value',
    figsize = (14,5)
)
plt.title("Nilai Index")
plt.ylabel("Total value")
plt.show()
```

```python
# ACF plot
plot_acf(
    df1['value'], 
    lags = 36
).set_size_inches((14,5));
```

```python
# PACF plot
plot_pacf(
     df1['value'], 
    lags = 36
).set_size_inches((14,5));
```

grafik visulaisasi diatas menunjukkan bahwa ACF dan PACF untuk deret waktu stasioner, masing-masing. Hasil viualisasi dari grafik diatas menunjukkan bahwa ACF tidak memiliki lag yang signifikan dan PACF memiliki konfident interval yang bergelombang / geometri ekponensial. Model ARMA
